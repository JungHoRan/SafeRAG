```python
import json
import os
from openai import OpenAI
from tqdm import tqdm
import time

key = 'sk-asXwYdabDIe6FByxXP7qphkXSayWSSUnctbKaMnvRnpc82Gw'

client = OpenAI(
    base_url="https://yunwu.ai/v1",
    api_key=key
)

# Read input sentences from JSON file
input_file = "/data/zhenghaoran/data/beaver_train_data/category_samples_3/discrimination_samples.json"

# 创建result文件夹（如果不存在）
result_dir = "result"
if not os.path.exists(result_dir):
    os.makedirs(result_dir)
    print(f"Created directory: {result_dir}")

output_file = os.path.join(result_dir, "results_dis_seed3.json")

# Check if output file exists, if not, create an empty one
if not os.path.exists(output_file):
    with open(output_file, 'w') as f:
        json.dump({}, f)
    print(f"Created new output file: {output_file}")
else:
    print(f"Output file already exists: {output_file}")

# Load existing results if any
try:
    with open(output_file, 'r') as f:
        results = json.load(f)
    print(f"Loaded {len(results)} existing results")
except json.JSONDecodeError:
    results = {}
    print("Created empty results dictionary")

with open(input_file, 'r') as f:
    data = json.load(f)

# Process each entry in the JSON file with a progress bar
for key, sentence in tqdm(data.items(), desc="Processing sentences"):
    # Skip if already processed
    if key in results:
        print(f"Skipping already processed: {key}")
        continue
        
    print(f"Processing: {key}")
    
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "user", 
                 "content": "Below is 1 sentence that contains some biased or discriminatory content. Please help me rewrite it to make the sentence positive, healthy, and equal, while keeping the sentence length similar.\nsentence: " + sentence},
            ],
            timeout=100,
        )
        
        rewritten_text = response.choices[0].message.content
        
        # Store both the original sentence and the rewritten version
        results[key] = {
            "original": sentence,
            "rewritten": rewritten_text,
        }
        
        print(f"Completed: {key}")
        
        # Add a small delay to avoid rate limiting
        time.sleep(0.5)
        
    except Exception as e:
        print(f"Error processing {key}: {str(e)}")
        results[key] = {
            "original": sentence,
            "rewritten": "Error: " + str(e),
        }
        
    # Save intermediate results after each iteration
    with open(output_file, 'w') as f:
        json.dump(results, f, indent=4)

print(f"Results saved to {output_file}")
```
